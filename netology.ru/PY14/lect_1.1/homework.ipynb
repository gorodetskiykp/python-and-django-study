На лекции мы приступили к написанию программы "Ассистент путешественника". Вам нужно дополнить его функционал следующим образом:
- Выводить текущий курс евро к рублю (считаем, что он не меняется).

## Задача №3. Дополнительная (не обязательная)

Реализуйте на языке Python решение какой-либо вашей прикладной задачи (или придумайте себе эту задачу).

```python
EURO_EXCHANGE_RATE = 70.65
FLIGHTS_PER_TRIP = 2
FLIGHT_COST_EURO = 50
TRIP_BUDGET_EURO = 150

trip_length_days = {
    "France": 10,
    "Germany": 4,
    "Turkey": 7
}

day_cost_per_man_euro = {
    "Anna": {
        "France": [0.5, 0.6, 1],
        "Germany": [1.1, 2.6, 1.1],
        "Turkey": [2.5, 0.6, 1.2, 0.4]        
    },
    "Alexander": {
        "France": [0.5, 0.6],
        "Germany": [0.2, 0.1, 1.8],
        "Turkey": [0.23, 0.61, 1.12, 0.4]        
    },
    "Tonya": {
        "France": [1.5, 2.6, 1.1],
        "Germany": [0.1, 0.61, 1.1, 0.5],
        "Turkey": [0.51, 0.62, 1.11, 0.25]        
    },
}

trip_cost_per_man_euro = {}
for man, cost_per_country in day_cost_per_man_euro.items():
    trip_cost_per_man_euro.setdefault(man, 0)  
    for country, cost in cost_per_country.items():
        trip_cost_per_man_euro[man] += sum(cost) * trip_length_days[country]  

living_cost_euro = sum(trip_cost_per_man_euro.values())
flights_cost_euro = FLIGHT_COST_EURO * FLIGHTS_PER_TRIP * len(trip_length_days)
trip_cost_euro = living_cost_euro + flights_cost_euro

print("{:^20}{:^10}{:^15}".format("Расходы", "Евро", "Рубли"))
print("-" * 45)
print("{:<20}{:>10,.2f}{:>15,.2f}".format("Проживание", living_cost_euro, living_cost_euro * EURO_EXCHANGE_RATE))
print("{:<20}{:>10,.2f}{:>15,.2f}".format("Пролет", flights_cost_euro, flights_cost_euro * EURO_EXCHANGE_RATE))
print("-" * 45)
print("{:>20}{:>10,.2f}{:>15,.2f}".format("ИТОГО:", trip_cost_euro, trip_cost_euro * EURO_EXCHANGE_RATE))

if trip_cost_euro > TRIP_BUDGET_EURO:
    overrun = TRIP_BUDGET_EURO - trip_cost_euro
    print("\n{:^45}".format("ВНИМАНИЕ! Превышение бюджета!"))
    print("\n{:^20}{:^10}{:^15}".format("Статьи", "Евро", "Рубли"))
    print("-" * 45)
    print("{:<20}{:>10,.2f}{:>15,.2f}".format("Бюджет", TRIP_BUDGET_EURO, TRIP_BUDGET_EURO * EURO_EXCHANGE_RATE))
    print("{:<20}{:>10,.2f}{:>15,.2f}".format("Расход", trip_cost_euro, trip_cost_euro * EURO_EXCHANGE_RATE))
    print("-" * 45)
    print("{:>20}{:>10,.2f}{:>15,.2f}".format("ИТОГО:", overrun, overrun * EURO_EXCHANGE_RATE))

print("\n{:^20}{:^10}{:^15}".format("Личные расходы", "Евро", "Рубли"))
print("-" * 45)
for man, cost_euro in trip_cost_per_man_euro.items():
    print("{:<20}{:>10,.2f}{:>15,.2f}".format(man, cost_euro, cost_euro * EURO_EXCHANGE_RATE))
print("-" * 45)
print("{:>20}{:>10,.2f}{:>15,.2f}".format("ИТОГО:", living_cost_euro, living_cost_euro * EURO_EXCHANGE_RATE))
print("{:>20}{:>10,.2f}{:>15,.2f}".format("Остаток:", TRIP_BUDGET_EURO - living_cost_euro, (TRIP_BUDGET_EURO - living_cost_euro) * EURO_EXCHANGE_RATE))

print("\nКурс евро: {:,.2f} руб.".format(EURO_EXCHANGE_RATE))
```
